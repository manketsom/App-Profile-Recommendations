```python
# Import necessary packages
import pandas as pd
```

```python
# Save the csv files and scan the databases
apple_store = pd.read_csv("AppleStore.csv")
google_playstore = pd.read_csv("googleplaystore.csv")
```

```python
# Print the columns of the databases
print(apple_store.columns)
print(google_playstore.columns)
# A lot of the column names need to be revised so that the user can better understand the contents of the column.
# This is more so evident for the apple_store dataset
google_playstore.rename(columns={"Content Rating" : "Content_Rating", "Last Updated" : " Last_Updated", "Current Ver" : " Current_Version",\
                                  "Android Ver" : "Android_Version"}, inplace=True)
apple_store.rename(columns={"rating_count_ver": "rating_count_version", "user_rating_ver" : "user_rating_version", "ver" : "version", \
                           "cont_rating" : "continued_rating", "rating_count_tot" : "rating_count_total"}, inplace=True)
# Drop any unnecessary columns that won't be used
apple_store.drop(columns= ["sup_devices.num", "ipadSc_urls.num", "lang.num", "vpp_lic"], inplace=True)
```
